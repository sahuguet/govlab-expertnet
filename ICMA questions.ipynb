# Import the important libraries

```python
import urllib2
import time
from bs4 import BeautifulSoup
from datetime import datetime
from datetime import timedelta
from dateutil.parser import parse
import sys
import json
import pandas as pd
import matplotlib.pyplot as plt
```

# Main function that process one page at a time, based on index value

```python
def getPage(index):
    html_doc = urllib2.urlopen('http://icma.org/en/icma/knowledge_network/questions/kn/Question/%d' % index).read()
    soup = BeautifulSoup(html_doc)
    return soup

def processDate(str):
    return parse(str).strftime("%d-%m-%Y")

# 7 days ago => a real data, based on today's date.
def processDateAgo(str):
    today = datetime.today()
    if 'hour' in str:
        return today.strftime("%d-%m-%Y")
    if 'day' in str:
        daysago = int(re.compile('(\d+) day[s]? ago').match(str).group(1))
        return (datetime.today() - timedelta(days=daysago)).strftime("%d-%m-%Y")
```

```python
def processItem(q_and_a):
    author = q_and_a.find('div', attrs={"class": "Attribution"}).a.text
    question_text = q_and_a.findAll('div', attrs={"class": "ListingBody"})[0].findAll('div')[0].text
    score = q_and_a.findAll('span', attrs={"class": "Score"})[0].text.strip()
    when = processDate(q_and_a.findAll('div', attrs={"class": "Dateline"})[0].text.strip())
    return { 'author': author, 'text': question_text, 'score': score, 'when': when }

def getNote(q):
    try:
        note = q.find('p', attrs={"class": "Note"}).strong.text.strip()
    except Exception, e:
        note = None
    return note

def processQuestion(s):
    q_and_a = s.findAll('div', attrs={"class": "Listing"})
    question = processItem(q_and_a[0])
    question['tags'] = [k.text.strip() for k in s.findAll('a', attrs={"class": "TagLink"})]
    __note__ = getNote(q_and_a[0])
    question['answer_by'] = processDate(__note__) if __note__ else None
    question['answers'] = []
    for answer in q_and_a[1:]:
        try:
            answer = processItem(answer)
            question['answers'].append(answer)
        except Exception, e:
            continue
    question['answers_count'] = len(question['answers'])
    return question
```

```python
FOLDER = 'ICMA_DATA'
questions = []
for i in range(23138, 23377):
#for i in range(22904, 23377):
#for i in range(22904, 22914):
    time.sleep(3)
    print >> sys.stderr, i
    sys.stderr.flush()
    page_soup = getPage(i)
    title = page_soup.title.text.split('|')[0].strip()
    if title == 'Questions' or title == 'Unauthorized':
        print >> sys.stderr, 'Invalid URL; skip'
        continue
    question = processQuestion(page_soup)
    question['title'] = title
    question['id'] = i
    questions.append(question)
    with open("%s/question-%d.json" % (FOLDER, i), 'w') as g:
        g.write(json.dumps(question, indent=2, sort_keys=True))
        g.close()
```

# Now that we have all the data, let's compute some stats.

```python
import glob
all_files = glob.glob(FOLDER + '/*.json')
all_data = []
for file in all_files:
    with open(file, 'rb') as f:
        data = json.load(f)
        f.close()
    all_data.append(data)
all_data_as_json = json.dumps(all_data)
```

How many questions get answered?

```python
all_questions = len(all_data)
questions_with_answer = len([k for k in all_data if k['answers_count'] > 0])
print "Percentage of questions with answers: %f." % (100.0 * questions_with_answer / all_questions)
```
